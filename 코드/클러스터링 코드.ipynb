```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('클러스터링 데이터2.csv', encoding='utf-8', index_col='행정동')
```

```python
data.head(3)
```

```python
data_copy = data.copy()
```

```python
# data_copy = data_copy.iloc[:, 1:]
```

```python
data_c = data_copy.values
```

```python
data_c
```

```python
# np.random.shuffle(data_c)
```

```python
# 스케일링 하는 게 좋은지 안 하는 게 좋은지는 지도로 찍어서 클러스터링 된 모양을 봐야 알 것 같음. 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

data_c = scaler.fit_transform(data_c)
```

```python
kmeans = KMeans(n_clusters=10, max_iter=100, verbose=1)
```

```python
clusters = kmeans.fit_predict(data_c)
```

```python
kmeans.labels_
```

```python
cluster_nums = []
for doc_num, cluster_num in enumerate(clusters): 
    cluster_nums.append(cluster_num)
    print(doc_num, cluster_num)
```

```python
cluster_nums
```

```python
data_copy['레이블'] = cluster_nums
```

```python
data_copy.head(20)
```

```python
pd.DataFrame(data_copy['레이블'])
```

```python
data_copy.to_csv('동별 클러스터링.csv')
```
